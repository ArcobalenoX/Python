```python
# coding = utf-8
import cv2
import os
import numpy as np
import shutil
import matplotlib.pyplot as plt
%matplotlib inline
```

# 读写

```python
im=cv2.imread("test.jpg")
im = cv2.cvtColor(im,cv2.COLOR_BGR2RGB)
plt.imshow(im)
```

```python
img_gray = cv2.cvtColor(im,cv2.COLOR_RGB2GRAY)
plt.imshow(img_gray)
```

```python
cv2.imwrite('gray.jpg',img_gray)
```

# fillConvexPoly 单体

```python
img = np.zeros((1080, 1920, 3), np.uint8)
triangle = np.array([[0, 0], [1000, 800], [500, 200]])
 
cv2.fillConvexPoly(img, triangle, (255, 200, 255))

plt.imshow(img)
plt.show()
```

# fillPoly 多体

```python
img = np.zeros((1080, 1920, 3), np.uint8)
area1 = np.array([[250, 200], [300, 100], [750, 800], [100, 1000]])
area2 = np.array([[1000, 200], [1500, 200], [1500, 400], [1000, 400]])
 
cv2.fillPoly(img, [area1, area2], (255, 255, 255),1)
 
plt.imshow(img)
plt.show()
```

# putText

```python
img = np.zeros((512,512,3), np.uint8)

font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(img,'OpenCV',(10,500), font, 4,(255,255,255),2,cv2.LINE_AA)

cv2.imshow('show',img)
plt.imshow(img)
```
