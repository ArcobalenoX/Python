```python
import matplotlib.pyplot as plt 
%matplotlib inline
```

# plot

```python
x = list(range(1,11))
y = [i**2 for i in x]
plt.plot(x,y,linewidth=5)
plt.title("square",fontsize=12)
plt.xlabel("x",fontsize=12)
plt.ylabel("y",fontsize=12)
plt.tick_params(axis='both',labelsize=10)
```

# scatter

```python
x = list(range(1,1000))
y = [i**2 for i in x]
plt.scatter(x,y,s=10,c='r',edgecolors='none')
plt.title("square",fontsize=12)
plt.xlabel("x",fontsize=12)
plt.ylabel("y",fontsize=12)
plt.tick_params(axis='both',labelsize=10)
plt.axis([0,100,0,10000])
```

```python
plt.scatter(x,y,c=y,cmap=plt.cm.Blues,s=40,edgecolors='none')
plt.title("square",fontsize=20)
plt.xlabel("x",fontsize=14)
plt.ylabel("y",fontsize=14)
plt.tick_params(axis='both',labelsize=10)
plt.savefig('square.jpg',bbox_inches='tight')
```

# random walk

```python
from random import choice
```

```python
class RandomWalk():
    """生成随机漫步数的类"""
    def __init__(self,num_points=5000):
        self.num_points =num_points
        self.x=[0]
        self.y=[0]
    def fill_walk(self):
        while len(self.x)<self.num_points :
            x_dir = choice([-1,1])
            x_dis = choice([0,1,2,3,4])
            x_step = x_dir * x_dis
            y_dir = choice([-1,1])
            y_dis = choice([0,1,2,3,4])
            y_step = y_dir * y_dis
            
            if x_step==0 and y_step==0:
                continue
            
            next_x = self.x[-1] + x_step
            next_y = self.y[-1] + y_step

            self.x.append(next_x)
            self.y.append(next_y)

```

```python
rw = RandomWalk()
rw.fill_walk()
plt.scatter(rw.x,rw.y,c=rw.y,cmap=plt.cm.Blues,s=10)
plt.scatter(0,0,c='g',s=20)
plt.scatter(rw.x[-1],rw.y[-1],c='r',s=20)
plt.axes().get_xaxis().set_visible(False)
plt.axes().get_yaxis().set_visible(False)
```

```python
rw = RandomWalk()
rw.fill_walk()
plt.scatter(rw.x,rw.y,c=rw.y,cmap=plt.cm.Blues,s=10)
plt.scatter(0,0,c='g',s=20)
plt.scatter(rw.x[-1],rw.y[-1],c='r',s=20)
plt.figure(dpi=128, figsize=(4,6))
plt.savefig('randomwalk.jpg',bbox_inches='tight')
```

```python
import pygal
from random import randint
```

```python
class Die():
    def __init__(self,sides=6):
        self.sides = sides
    def roll(self):
        return randint(1,self.sides)
```

```python
d6 = Die(6)
results = []
for roll_num in range(100):
    results.append(d6.roll())
frequencies = []
for v in range(1,d6.sides+1):
    frequencies.append(results.count(v))    
```

```python
hist = pygal.Bar()
hist.title='roll die6 1000'
hist.x_labels = ['1','2','3','4','5','6']
hist.x_title = 'result'
hist.y_title = 'frequency'
hist.add('D6',frequencies)
hist.add('D7',frequencies)
hist.render_to_file('die6.svg')
```

```python
hist = pygal.Line()
hist.title='roll die6 1000'
hist.x_labels = ['1','2','3','4','5','6']
hist.x_title = 'result'
hist.y_title = 'frequency'
hist.add('D6',frequencies)
hist.add('D7',frequencies)
hist.render_to_file('die6.svg')
```
